```python
import pandas as pd
```

```python
# Load the data
df = pd.read_csv('annual_time_series_with_airports&airlines.csv')

# Perform groupby operation and calculate the yearly totals
grouped = df.groupby(['ORIGIN_AIRPORT_ID', 'AIRCRAFT_TYPE'])['2019', '2020', '2021', '2022'].sum().reset_index()

# Drop duplicate rows for non-aggregate columns and keep the first one
non_aggregate = df[['ORIGIN_AIRPORT_ID', 'AIRCRAFT_TYPE', 'ORIGIN_STATE_ABR', 'CODE', 'NAME', 'Description']].drop_duplicates(subset=['ORIGIN_AIRPORT_ID', 'AIRCRAFT_TYPE'])

# Merge the grouped dataframe with the non-aggregate dataframe
merged = pd.merge(non_aggregate, grouped, on=['ORIGIN_AIRPORT_ID', 'AIRCRAFT_TYPE'])

# Reorder the columns
merged = merged[['ORIGIN_AIRPORT_ID', 'ORIGIN_STATE_ABR', 'CODE', 'NAME', 'AIRCRAFT_TYPE', 'Description', '2019', '2020', '2021', '2022']]

# Save the output to a single file
merged.to_csv('annual_time_series_actypes@airports.csv', index=False)
```

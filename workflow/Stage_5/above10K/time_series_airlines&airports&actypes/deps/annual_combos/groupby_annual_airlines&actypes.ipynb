```python
import pandas as pd
```

```python
# Load the data
df = pd.read_csv('annual_time_series_with_airports&airlines.csv')

# Perform groupby operation and calculate the yearly totals
grouped = df.groupby(['AIRLINE_ID', 'AIRCRAFT_TYPE'])['2019', '2020', '2021', '2022'].sum().reset_index()

# Drop duplicate rows for non-aggregate columns and keep the first one
non_aggregate = df[['AIRLINE_ID', 'AIRCRAFT_TYPE', 'ORIGIN_STATE_ABR', 'CODE', 'NAME', 'UNIQUE_CARRIER_NAME', 'Description']].drop_duplicates(subset=['AIRLINE_ID', 'AIRCRAFT_TYPE'])

# Merge the grouped dataframe with the non-aggregate dataframe
merged = pd.merge(non_aggregate, grouped, on=['AIRLINE_ID', 'AIRCRAFT_TYPE'])

# Reorder the columns
merged = merged[['AIRLINE_ID', 'UNIQUE_CARRIER_NAME', 'AIRCRAFT_TYPE', 'Description', '2019', '2020', '2021', '2022']]

# Save the output to a single file
merged.to_csv('annual_time_series_airlines&actypes.csv', index=False)


```

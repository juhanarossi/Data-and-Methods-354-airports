```python
df = pd.read_csv('quarterly_deps_per_cluster_above10K.csv')
```

```python
df.columns
```

```python
quarter_cols = ['1Q2019', '2Q2019', '3Q2019', '4Q2019', '1Q2020', '2Q2020', '3Q2020',
       '4Q2020', '1Q2021', '2Q2021', '3Q2021', '4Q2021', '1Q2022', '2Q2022',
       '3Q2022', '4Q2022']

output_df = df[['category']].copy() # start with 'category' column

for i in range(1, len(quarter_cols)):
    # Add current quarter column
    output_df[quarter_cols[i]] = df[quarter_cols[i]]
    
    # Calculate the change from previous quarter
    output_df['change_from_' + quarter_cols[i-1] + '_to_' + quarter_cols[i]] = df[quarter_cols[i]] - df[quarter_cols[i-1]]

    # Calculate the percent change from previous quarter
    pct_change = df[quarter_cols[i]] / df[quarter_cols[i-1]] - 1
    output_df['pct_change_from_' + quarter_cols[i-1] + '_to_' + quarter_cols[i]] = round(pct_change * 100, 1) # Multiply by 100 and round to 1 decimal place

output_df[quarter_cols[0]] = df[quarter_cols[0]] # Add the first quarter column at the end




```

```python
output_df.to_csv('quarterly_deps_per_cluster_above10K_with_changes.csv', index=False)
```

```python
output_df
```

```python
output2_df = output_df[['category']].join(output_df.filter(like='pct_change_from_'))
```

```python
output2_df
```

```python
# Prepare a dictionary for renaming
rename_dict = {col: col.split('_to_')[-1] for col in output2_df.columns if 'pct_change_from_' in col}

# Retain the 'category' column
if 'category' in output2_df.columns:
    rename_dict['category'] = 'category'

# Rename columns
output2_df.rename(columns=rename_dict, inplace=True)
```

```python
output2_df
```

```python
output2_df.to_csv('quarterly_deps_per_cluster_above10K_with_pct_change_only.csv', index=False)
```

```python
quarter_cols = ['1Q2019', '2Q2019', '3Q2019', '4Q2019', '1Q2020', '2Q2020', '3Q2020',
       '4Q2020', '1Q2021', '2Q2021', '3Q2021', '4Q2021', '1Q2022', '2Q2022',
       '3Q2022', '4Q2022']

year_on_year_df = df[['category']].copy() # start with 'category' column

# Notice that we start from index 4 (representing 1Q2020) because we want to compare it with the same quarter of the previous year.
for i in range(4, len(quarter_cols)):
    # Add current quarter column
    year_on_year_df[quarter_cols[i]] = df[quarter_cols[i]]
    
    # Calculate the change from the same quarter of the previous year
    year_on_year_df['change_between_' + quarter_cols[i-4] + '_and_' + quarter_cols[i]] = df[quarter_cols[i]] - df[quarter_cols[i-4]]

    # Calculate the percent change from the same quarter of the previous year
    pct_change = df[quarter_cols[i]] / df[quarter_cols[i-4]] - 1
    year_on_year_df['pct_change_between_' + quarter_cols[i-4] + '_and_' + quarter_cols[i]] = round(pct_change * 100, 1) # Multiply by 100 and round to 1 decimal place

for col in quarter_cols[:4]: # Add the first 4 quarter columns at the end
    year_on_year_df[col] = df[col]
```

```python
year_on_year_df
```

```python
year_on_year_df.to_csv('quarterly_deps_per_cluster_above10K_with_YonY_change.csv', index=False)
```

```python
year_on_year2_df = year_on_year_df[['category']].join(year_on_year_df.filter(like='pct_change_between_'))
```

```python
year_on_year2_df
```

```python
# Prepare a dictionary for renaming
rename2_dict = {col: col.split('_and_')[-1] for col in year_on_year2_df.columns if 'pct_change_between_' in col}

# Retain the 'category' column
if 'category' in year_on_year2_df.columns:
    rename2_dict['category'] = 'category'

# Rename columns
year_on_year2_df = year_on_year2_df.rename(columns=rename2_dict)
```

```python
year_on_year2_df.to_csv('quarterly_deps_per_cluster_above10K_with_YonY_pct_change_only.csv', index=False)
```

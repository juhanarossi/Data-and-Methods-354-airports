```python
df = pd.read_csv('quarterly_pax_clusters_above10K.csv')
```

```python
#df.drop(columns=['Unnamed: 0'], inplace=True)
df.to_csv('quarterly_deps_clusters_above10K.csv', index=False)
```

```python
df
```

```python
# Define the quarterly columns
annum_cols = ['total_2019', 'total_2020', 'total_2021', 'total_2022']

# Group the DataFrame by 'category' and calculate the total for each quarter
df2 = df.groupby('category')[annum_cols].sum()

# Reset the index to make 'category' a column again
df2 = df2.reset_index()
```

```python
df2
```

```python
# Define the specific order for the 'category' column
cat_order = ['more than 5 million', '1-5 million', '100k-1 million', '10k-100k', 'less than 10k']

# Convert the 'category' column to a categorical type with the specified order
df2['category'] = pd.Categorical(df2['category'], categories=cat_order, ordered=True)

# Sort by 'category'
df2 = df2.sort_values('category')
```

```python
df2.to_csv('annual_deps_per_cluster_above10K.csv', index=False)
```

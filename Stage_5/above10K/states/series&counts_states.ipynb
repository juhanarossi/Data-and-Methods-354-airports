```python
import pandas as pd
```

```python
df = pd.read_csv('annual_deps_clusters_above10K.csv')
```

```python
df
```

```python
#df.columns
```

```python
# Perform value counts on the 'category' column
cat_counts = df['category'].value_counts().reset_index()

# Rename the columns
cat_counts.columns = ['category', 'count']
```

```python
cat_counts
```

```python
# Define the custom order
order = ['more than 5 million', '1-5 million', '100k-1 million', '10k-100k', 'less than 10k']

# Set 'category' as the index of cat_counts
cat_counts = cat_counts.set_index('category')

# Reorder the rows based on the custom order
cat_counts = cat_counts.loc[order]

# Reset the index
cat_counts = cat_counts.reset_index()
```

```python
cat_counts.to_csv('cat_counts.csv', index=False)
```

```python
# Group by 'ORIGIN_STATE_ABR' and 'ORIGIN_AIRPORT_ID', and then sum the totals
output_df = df.groupby(['ORIGIN_STATE_ABR'])['total_2019', 'total_2020', 'total_2021', 'total_2022'].sum().reset_index()

# Rename the columns
output_df = output_df.rename(columns={
    'total_2019': '2019',
    'total_2020': '2020',
    'total_2021': '2021',
    'total_2022': '2022',
})
```

```python
output_df
```

```python
output_df['pct_change'] = ((output_df['2022'] - output_df['2019']) / output_df['2019']) * 100

# To round to one decimal place
output_df['pct_change'] = output_df['pct_change'].round(1)
```

```python
output_df
```

```python
output_df.to_csv('annual_deps_totals_states.csv', index=False)
```

```python
# Group by 'ORIGIN_STATE_ABR' and count the number of 'ORIGIN_AIRPORT_ID' for each 'ORIGIN_STATE_ABR'
airport_count = df.groupby('ORIGIN_STATE_ABR')['ORIGIN_AIRPORT_ID'].size().reset_index()

# Rename the columns
airport_count = airport_count.rename(columns={
    'ORIGIN_STATE_ABR': 'state',
    'ORIGIN_AIRPORT_ID': 'airport_count',
})
```

```python
airport_count
```

```python
# Create a pivot table
airport2_df = df.pivot_table(index='ORIGIN_STATE_ABR', columns='category', aggfunc='size', fill_value=0).reset_index()

# Rename the index column
airport2_df = airport2_df.rename(columns={'ORIGIN_STATE_ABR': 'state'})
```

```python
airport2_df
```

```python
# Merge airport_df with airport2_df
merged_df = airport_count.merge(airport2_df, on='state', how='left')
```

```python
merged_df
```

```python
# Reorder the columns
merged_df = merged_df[['state', 'airport_count', 'more than 5 million', '1-5 million', '100k-1 million', '10k-100k', 'less than 10k']]
```

```python
merged_df
```

```python
merged_df.to_csv('airports_and_categories_@states.csv', index=False)
```

```python
# Create a pivot table
airport3_df = df.pivot_table(index='ORIGIN_STATE_ABR', columns='change_category', aggfunc='size', fill_value=0).reset_index()

# Rename the index column
airport3_df = airport3_df.rename(columns={'ORIGIN_STATE_ABR': 'state'})
```

```python
airport3_df
```

```python
merged2_df = airport_count.merge(airport3_df, on='state', how='left')
```

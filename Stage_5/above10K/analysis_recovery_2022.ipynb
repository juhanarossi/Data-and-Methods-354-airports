```python
import pandas as pd
```

```python
df = pd.read_csv('annual_deps_clusters_above10K.csv')
```

```python
df.head()
```

```python
# Create the 'diff_2022' column
df['diff_2022'] = df['total_2022'] - df['total_2021']

# Create the 'change_type' column
df['change_type'] = 'unchg'  # default value
df.loc[df['total_2022'] > df['total_2021'], 'change_type'] = 'more'
df.loc[df['total_2022'] < df['total_2021'], 'change_type'] = 'less'
```

```python
df.head()
```

```python
df.to_csv('annual_deps_clusters_above10K.csv', index=False)
```

```python
# Group by 'category' and get value counts for 'change_type'
new_df = df.groupby('category')['change_type'].value_counts().unstack().reset_index()
```

```python
new_df.head()
```

```python
new_df.to_csv('annual_deps_change_type_count.csv', index=False)
```

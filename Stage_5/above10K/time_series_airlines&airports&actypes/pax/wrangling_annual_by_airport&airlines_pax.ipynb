```python
# Define the list of years
years = ['2019', '2020', '2021', '2022']

# List of months
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Initial dataframe
output_df = None

for year in years:
    # Load the data
    filename = f"time_series_pax_by_airport&airline_{year}_@above10K.csv"
    df = pd.read_csv(filename)

    # Calculate the annual total and create a new dataframe with it
    df[year] = df[months].sum(axis=1)
    df_yearly = df.drop(months + ['total'], axis=1)

    # Merge with the output dataframe
    if output_df is None:
        output_df = df_yearly
    else:
        output_df = pd.merge(output_df, df_yearly, how='outer', on=['ORIGIN_AIRPORT_ID', 'CODE', 'NAME', 'AIRLINE_ID', 'UNIQUE_CARRIER_NAME', 'AIRCRAFT_TYPE', 'Description'])

# Save the output to a single file
output_df.to_csv('annual_time_series_pax_with_airports&airlines.csv', index=False)

```

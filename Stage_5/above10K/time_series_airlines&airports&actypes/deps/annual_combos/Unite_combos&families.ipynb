```python
# Step 1: Calculate total for 2022
total_2022 = families_df['2022'].sum()

# Step 2: Filter data for specific aircraft families and calculate 2022 totals
specific_families = ['Airbus A319/A320', 'Airbus A321', 'Boeing 737', 'Boeing 737 Max']
specific_totals = families_df[families_df['AIRCRAFT_FAMILY'].isin(specific_families)].groupby('AIRCRAFT_FAMILY')['2022'].sum()

# Step 3: Calculate total for 'OTHER'
other_total = total_2022 - specific_totals.sum()

# Step 4: Calculate percentage shares
percentage_shares = (specific_totals / total_2022) * 100
percentage_shares.loc['OTHER'] = (other_total / total_2022) * 100

# Convert to dataframe and reset index
percentage_shares_df = percentage_shares.reset_index()
percentage_shares_df.columns = ['AIRCRAFT_FAMILY', '2022_PERCENTAGE_SHARE']

print(percentage_shares_df)
```

```python
percentage_shares_df.to_csv('pct_shares_by_acfamily_in_2022.csv', index=False)
```

```python
# Step 1: Calculate total for 2022
total_2019 = families_df['2019'].sum()

# Step 2: Filter data for specific aircraft families and calculate 2022 totals
specific_families = ['Airbus A319/A320', 'Airbus A321', 'Boeing 737', 'Boeing 737 Max']
specific_totals = families_df[families_df['AIRCRAFT_FAMILY'].isin(specific_families)].groupby('AIRCRAFT_FAMILY')['2019'].sum()

# Step 3: Calculate total for 'OTHER'
other_total = total_2019 - specific_totals.sum()

# Step 4: Calculate percentage shares
percentage_shares = (specific_totals / total_2019) * 100
percentage_shares.loc['OTHER'] = (other_total / total_2019) * 100

# Convert to dataframe and reset index
percentage_shares_df = percentage_shares.reset_index()
percentage_shares_df.columns = ['AIRCRAFT_FAMILY', '2019_PERCENTAGE_SHARE']

print(percentage_shares_df)
```

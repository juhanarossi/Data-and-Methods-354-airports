```python
# Load the data
df = pd.read_csv('annual_time_series_with_airports&airlines.csv')

# Perform groupby operation and calculate the yearly totals
grouped = df.groupby(['ORIGIN_AIRPORT_ID', 'AIRLINE_ID'])['2019', '2020', '2021', '2022'].sum().reset_index()

# Drop duplicate rows for non-aggregate columns and keep the first one
non_aggregate = df[['ORIGIN_AIRPORT_ID', 'AIRLINE_ID', 'ORIGIN_STATE_ABR', 'CODE', 'NAME', 'UNIQUE_CARRIER_NAME']].drop_duplicates(subset=['ORIGIN_AIRPORT_ID', 'AIRLINE_ID'])

# Merge the grouped dataframe with the non-aggregate dataframe
merged = pd.merge(non_aggregate, grouped, on=['ORIGIN_AIRPORT_ID', 'AIRLINE_ID'])

# Reorder the columns
merged = merged[['ORIGIN_AIRPORT_ID', 'ORIGIN_STATE_ABR', 'CODE', 'NAME', 'AIRLINE_ID', 'UNIQUE_CARRIER_NAME', '2019', '2020', '2021', '2022']]

# Save the output to a single file
merged.to_csv('time_series_by_airlines@airports.csv', index=False)

```

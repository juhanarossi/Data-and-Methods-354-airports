```python
df1 = pd.read_csv('merged_airport_data.csv')
```

```python
df1.head()
```

```python
df2 = pd.read_csv('above10K_airports_list.csv')
```

```python
name_df = df2.loc[:, ['ORIGIN_AIRPORT_ID', 'NAME']]
```

```python
name_df.head()
```

```python
df1.columns
```

```python
# Reset index if 'ORIGIN_AIRPORT_ID' is an index in either DataFrame
df1.reset_index(inplace=True)
name_df.reset_index(inplace=True)

# Then, rename the 'NAME' column in 'name_df' to 'NAME_DEST'
name_df.rename(columns={'NAME': 'NAME_DEST'}, inplace=True)

# Perform the merge operation
df1 = pd.merge(df1, name_df, 
                     left_on='DEST_AIRPORT_ID', right_on='ORIGIN_AIRPORT_ID', how='left')

# Now, drop the 'DEST_CITY_NAME' column
df1.drop('DEST_CITY_NAME', axis=1, inplace=True)

# Rearrange the columns to your specified order
df1 = df1[['ORIGIN_AIRPORT_ID_x', 'CODE', 'ORIGIN_STATE_ABR', 'NAME', 'LOCATION',
                       'category', 'EAS', 'hub_cat', 'diff', 'pct_change', 'change_category', 
                       'DEST_AIRPORT_ID', 'DEST', 'hub_cat_dest', 'NAME_DEST', 
                       'DEST_STATE_ABR', '2019', '2020', '2021', '2022']]
```

```python
df1.head()
```

```python
df1.rename(columns={'ORIGIN_AIRPORT_ID_x': 'ORIGIN_AIRPORT_ID'}, inplace=True)
```

```python
df1.head()
```

```python
df1.to_csv('annual_time_series_deps_by_origin&dest.csv', index=False)
```
